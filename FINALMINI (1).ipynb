# **Test statistic**

### The significance level of p-value = 0.05

# **T-test**

```python
from scipy import stats

red_wines = df[df['Id'] < 1000]['alcohol']
white_wines = df[df['Id'] >= 1000]['alcohol']

t_statistic, p_value = stats.ttest_ind(red_wines, white_wines)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

### Interpretation: Since the p-value is greater than chosen significance level, can conclude whether there's a significant difference in alcohol levels.

# **Z-test**

```python
from statsmodels.stats.weightstats import ztest

z_statistic, p_value = ztest(df['fixed acidity'], value=7)
print("Z-statistic:", z_statistic)
print("P-value:", p_value)
```

Interpretation: Since the p-value is greater than chosen significance level, can conclude whether there's a significant difference in alcohol levels.

# **ANNOVA**

```python
from scipy.stats import f_oneway

quality_5 = df[df['quality'] == 5]['alcohol']
quality_6 = df[df['quality'] == 6]['alcohol']
quality_7 = df[df['quality'] == 7]['alcohol']

f_statistic, p_value = f_oneway(quality_5, quality_6, quality_7)
print("F-statistic:", f_statistic)
print("P-value:", p_value)
```

Interpretation: Since the p-value is greater than chosen significance level, can conclude whether there's a significant difference in alcohol levels.


# **Chi-Square Test**

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['quality'], df['Id'] < 1000)
chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
print("Chi-squared statistic:", chi2_stat)
print("P-value:", p_value)
```

Interpretation: Since the p-value is less than your chosen significance level, you can conclude whether there's an association between wine quality ).
